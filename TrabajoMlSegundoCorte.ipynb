```python
#seccion de importacion de la librerias que utilizaremos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importar el dataset y guardarlo en la variables df
df = pd.read_csv('H:/dataset2.csv')
```

```python
#comprobamos si los datos de importaron de manera exitosa imprimiendo los 5 valoes
df.head(5)
```

```python
#identificar si hay datos duplicados
df.duplicated()
```

```python
#eliminar datos duplicados
df.drop_duplicates()
```

```python
#Eliminamos las columnas que no nos interesa por ahora
df = df.drop(["Region", "Country", "Country","Item Type","Sales Channel","Order Priority","Order Date","Order ID","Ship Date"	], axis=1)
```

```python
#validamos si las filas se borraron exitosamente
df.head(1)
```

```python
#¡reemplazamos los valores faltante con el valor siguiente (o anterior) en el marco de datos con el siguiente funcion!
df.fillna(method = "ffill")
```

```python
df2=df
```

```python
#normalizaremos los datos
from sklearn import preprocessing

x = df2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df2=pd.DataFrame(x_scaled, columns=df.columns)
```

```python
#mostramos los 5 primeros registro ya realizados
df2.head(5)
```

```python
#guardamos los datos normalizado en csv
df2.to_csv('H:/normalizados.csv', header=False, index=False)
```
